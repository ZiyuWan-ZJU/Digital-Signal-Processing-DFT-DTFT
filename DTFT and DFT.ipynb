```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
import numpy as np
from scipy import signal
```

```python
fs=2*np.pi
def DTFT(x,a,b):
    num =np.array(x)
    den = np.array([1])
    ws, h = signal.freqz(num,a=den,whole=True,worN=np.linspace(a,b,512))
    plt.plot(ws,np.abs(h))
    plt.title('DTFT Spectrum')
    return 

def FFT(x,N):
    xs=np.linspace(0,fs,N+1)
    plt.scatter(xs[0:N],np.abs(np.fft.fft(x,n=N)))
    num =np.array(x)
    den = np.array([1])
    ws, h = signal.freqz(num,a=den,whole=True,worN=np.linspace(0,fs,512))
    plt.plot(ws,np.abs(h))
    plt.xlabel('Frequancy $\Omega$')
    plt.ylabel('Amplitude')
    plt.title('%i-point DFT'%N)
    return np.fft.fft(x,n=N)
```

```python
x=np.arange(0,16,1)  
y=[np.cos(math.pi*i/2) for i in x]  
type(y)
y=np.array(y)
plt.plot(x,y)
```

```python
DTFT(y,-fs/2,fs/2)
```

```python
FFT(y,16)
```
